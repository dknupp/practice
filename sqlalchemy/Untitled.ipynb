```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import (create_engine,
                        Column,
                        Integer,
                        String)

engine = create_engine('sqlite:///:memory:', echo=True)
Base = declarative_base()
```

```python
class User(Base):
    __tablename__ = 'users'
    
    id = Column(Integer, primary_key=True)
    name = Column(String)
    fullname = Column(String)
    password = Column(String)
    
    def __repr__(self):
        return "<User(name='{}', fullname='{}', password='{}')>" \
               .format(self.name, self.fullname, self.password)        
```

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

s = 'BANANA'
def make_words(s, starts_with_vowel=True):
    count = 0
    for i in range(len(s)):
        if starts_with_vowel:
            if s[i] not in 'AEIOU':
                continue
        else:  # must start with a consonant
            if s[i] in 'AEIOU':
                continue                
        count += len(s[i:])
    return count

player1, score1 = 'Stuart', make_words(s, False)
player2, score2 = 'Kevin', make_words(s, True)

if score1 > score2:
    print player1, score1
elif score2 > score1:
    print player2, score2
else:
    print 'Draw'
```
